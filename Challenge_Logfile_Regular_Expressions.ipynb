```python
import re
import pprint

def extract_endpoint(log_line, log_format):
    """
    Extract the endpoint from a log line using the specified log format.
    """
    if log_format == "format1":
        match = re.search(r'"(GET|POST|HEAD|PUT|DELETE|CONNECT|OPTIONS|TRACE) ([^"]*)"', log_line)
        if match:
            return match.group(2)
    elif log_format == "format2":
        match = re.search(r'"([^"]*)"', log_line)
        if match:
            return match.group(1)
    # Add additional log formats as needed
    else:
        return None

def check_logs_for_404(log_file, log_format):
    """
    Check the log file for 404 errors and extract the endpoints.
    """
    status_code_counter = {}
    endpoint_counter = {}

    with open(log_file, "r") as f:
        logs = f.readlines()

    for log in logs:
        match = re.search(r'"([^"]*)" ([0-9]{3})', log)
        if match:
            status_code = match.group(2)
            if status_code == "404":
                endpoint = extract_endpoint(log, log_format)
                if endpoint:
                    if endpoint in endpoint_counter:
                        endpoint_counter[endpoint] += 1
                    else:
                        endpoint_counter[endpoint] = 1
                if status_code in status_code_counter:
                    status_code_counter[status_code] += 1
                else:
                    status_code_counter[status_code] = 1

    pprint.pprint(f"Endpoint counter: {endpoint_counter}")
    
    pprint.pprint(f"Status code counter: {status_code_counter}")
    

# Usage example
check_logs_for_404("access.log", "format1")
```
